problem link: https://leetcode.com/problems/find-minimum-in-rotated-sorted-array/description/

```python
from typing import List
```

```python
class Solution:
    def findMin(self, nums: List[int]) -> int:
        res = nums[0]
        l, r = 0, len(nums) - 1

        while l <= r:
            if nums[l] < nums[r]:
                res = min(res, nums[l])
                break

            m = (l + r) // 2
            res = min(res, nums[m])
            if nums[m] >= nums[l]:
                l = m + 1
            else:
                r = m - 1
        return res
```

```python
solution = Solution()
```

```python
nums = [5,1,2,3,4]
```

```python
print(solution.findMin(nums))
```
