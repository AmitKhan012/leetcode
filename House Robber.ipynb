```python
from typing import List
```

```python
class Solution:
  def rob(self, nums: List[int]) -> int:
    rob1, rob2 = 0, 0

    for n in nums:
      temp = max(n + rob1, rob2)
      rob1 = rob2
      rob2 = temp
    return rob2
```

```python
solution = Solution()
```

```python
nums = [2,7,9,3,1]
```

```python
output = solution.rob(nums)
print(output)
```
